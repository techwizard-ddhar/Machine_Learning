```python
movie_matrix = df.pivot_table(index = "user_id", columns = "movie_names", values = "rating") ## Imp
```

```python
movie_matrix
```

## Movie Recommendation

```python
starwars_user_rating = movie_matrix["Star Wars (1977)"]
starwars_user_rating
```

```python
starwars_user_rating.dropna(inplace=True)
```

```python
starwars_user_rating.head()
```

```python
similiar_to_starwars = movie_matrix.corrwith(starwars_user_rating)
```

```python
similiar_to_starwars
```

```python
similiar_to_starwars.dropna(inplace=True)
```

```python
similiar_to_starwars
```

```python
recomm = pd.DataFrame(similiar_to_starwars, columns=['Correlation'])
```

```python
recomm
```

```python
recomm.sort_values(by='Correlation', ascending=False)
```

```python
rating.head()
```

```python
recomm = recomm.join(rating['num of ratings'])  # Imp ---> join
```

```python
recomm
```

```python
recomm[recomm["num of ratings"]>100].sort_values("Correlation", ascending=False)
```

## Predict Function

```python
def predict(movie_name):
    movie_user_rating = movie_matrix[movie_name]
    similiar_to_movie = movie_matrix.corrwith(movie_user_rating)
    
    recomm = pd.DataFrame(similiar_to_movie, columns=['Correlation'])
    recomm.dropna(inplace=True)
    
    recomm = recomm.join(rating['num of ratings'])
    result=recomm[recomm["num of ratings"]>100].sort_values("Correlation", ascending=False)
    return result

prediction = predict("Titanic (1997)")
print(prediction)
```
