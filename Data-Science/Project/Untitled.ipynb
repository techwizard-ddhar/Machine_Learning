```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python

plt.scatter(x=data['LotArea'], y=data['SalePrice'], c=data['MSZoning'].astype('category').cat.codes)
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Lot Area vs Sale Price')
plt.colorbar(label='MSZoning')
plt.grid(True)
plt.show()
```
