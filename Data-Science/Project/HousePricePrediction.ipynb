```python
import pandas as pd
import numpy as np

def step_gradient(points, learning_rate, m):
    num_features = len(m)
    m_slope = np.zeros(num_features)
    N = len(points)  # No. of rows
    for i in range(N):
        x = points.iloc[i, :-1]  
        y = points.iloc[i, -1]  
        for j in range(num_features):
            m_slope[j] += (-2/N) * (y - np.dot(m, x)) * x[j]
    m = m - learning_rate * m_slope  # Update m here
    return m


def gradient_descent(points, learning_rate, num_iterations):
    num_features = points.shape[1] - 1
    m = np.zeros(num_features)
    for i in range(num_iterations):
        m = step_gradient(points, learning_rate, m)
        print(i, "Cost :", cost(points, m))
    return m

def cost(points, m):
    total_cost = 0
    N = len(points)
    for i in range(N):
        x = points.iloc[i, :-1]
        y = points.iloc[i, -1]
        total_cost += (1/N) * (y - np.dot(m, x)) ** 2
    return total_cost

def run():
    df = pd.read_csv("train.csv")
    df = df.select_dtypes(include=['float64', 'int64'])
    
    learning_rate = 0.01
    num_iterations = 10
    m = gradient_descent(df, learning_rate, num_iterations)
    print("Theta:", m)

run()
```
