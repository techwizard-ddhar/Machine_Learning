```python
import numpy as np
```

```python
df = np.loadtxt("data.csv", delimiter=",")
```

```python
df # 2D array
```

```python
x = df[:,0]
y = df[:,1]
```

```python
x.shape # 1D array
```

```python
y.shape
```

```python
from sklearn import model_selection # test = 25%; train = 75% {Default}
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x,y)
```

```python
def fit(X_train, Y_train):
    num = (X_train*Y_train).mean() - X_train.mean()*Y_train.mean()
    den = (X_train**2).mean() - X_train.mean()**2
    m = num/den
    c = Y_train.mean() - m*X_train.mean()
    return(m,c)
```

```python
def predict(X, m, c):
    y_predict = m*X + c
    return y_predict

def score(y_truth, y_predict):
    u = ((y_truth - y_predict) ** 2).sum()
    v = ((y_truth - y_truth.mean()) ** 2).sum()
    return 1 - u / v

def cost(x,y,m,c):
    return ((y-(m * x + c))**2).mean()
```

```python
m,c = fit(X_train, Y_train)

# test data
y_test_score = predict(X_test, m, c)
print("Test Score:",score(Y_test, y_test_score))

# train data
y_train_score = predict(X_train, m, c)
print("Train Score:",score(Y_train, y_train_score))

print("M value, C value :", m, c) 
print("Cost on training data:", cost(X_train, Y_train, m, c))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Train

x_line = np.arange(20,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
plt.scatter(X_train, Y_train)
plt.grid(True)
plt.show()
```

```python
#Test

x_line = np.arange(20,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "b")
plt.scatter(X_test, Y_test)
plt.grid(True)
plt.show()
```

```python
# Predictions on test data
y_test_pred = predict(X_test, m, c)

# Predictions on training data
y_train_pred = predict(X_train, m, c)

# Compare actual vs. predicted for test data
plt.scatter(Y_test, y_test_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs. Predicted (Test Data)")
plt.show()

# Compare actual vs. predicted for training data
plt.scatter(Y_train, y_train_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs. Predicted (Training Data)")
plt.show()
```
