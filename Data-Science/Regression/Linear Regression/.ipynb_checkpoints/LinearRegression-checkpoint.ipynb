```python
import numpy as np
```

```python
data = np.loadtxt("data.csv", delimiter=",")
```

```python
data.shape
```

```python
x = data[:,0].reshape(-1, 1)
y = data[:,1]
```

```python
x.shape
```

```python
y.shape
```

```python
pip install scikit-learn
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x,y) #split provides random number and it changes everytime
```

```python
from sklearn.linear_model import LinearRegression
algo1 = LinearRegression() # creating objects
algo1.fit(X_train,Y_train)
```

```python
x = data[:,0].reshape(-1,1) # ValueError: Expected 2D array, got 1D array instead:

y = data[:,1]
```

```python
algo1.coef_
```

```python
algo1.intercept_
```

```python
import matplotlib.pyplot as plt
```

```python
# TRAIN

m = algo1.coef_
c = algo1.intercept_

x_line = np.arange(20,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
plt.scatter(X_train, Y_train)
plt.grid(True)
plt.show()
```

```python
#TEST

plt.plot(x_line, y_line, "r")
plt.scatter(X_test, Y_test)
plt.grid(True)
plt.show()
```

## Coefficient of determination



```python
score_test = algo1.score(X_test, Y_test)
score_train = algo1.score(X_train, Y_train)
score_test, score_train
```

```python
# Predictions on test data
y_test_pred = algo1.predict(X_test)

# Predictions on training data
y_train_pred = algo1.predict(X_train)

# Compare actual vs. predicted for test data
plt.scatter(Y_test, y_test_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs. Predicted (Test Data)")
plt.show()

# Compare actual vs. predicted for training data
plt.scatter(Y_train, y_train_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs. Predicted (Training Data)")
plt.show()
```
