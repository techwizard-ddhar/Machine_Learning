```python
import pandas as pd
```

```python
a = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
a
```

```python
df = a.copy()
```

```python
df.columns = ["sepal-length", "sepal-width", "petal-length", "petal-width", "flower-type"] 
```

```python
df.head()
```

```python
df["diff_sw_pw"] = df["sepal-width"] - df["petal-width"]
```

```python
df["Gender"] = "Female"
```

```python
df.iloc[0:10,6] = "Male"
df.head()
```

```python
def f(s):
    if s=="Male":
        return 0
    else:
        return 1
    
df["sex"] = df["Gender"].apply(f)
df.head()
```
