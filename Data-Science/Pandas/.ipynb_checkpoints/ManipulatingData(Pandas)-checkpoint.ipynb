```python
import pandas as pd
```

```python
a = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
a
```

```python
df = a.copy()
```

```python
df.columns = ["sepal-length", "sepal-width", "petal-length", "petal-width", "flower-type"] 
```

```python
df.head()         # (---- = (sepal-length,...,flower-type) = column headers) and (| = (0,1,2,....) = labels for rows) 
```

```python
df.drop(0)     # makes a copy of the original DataFrame without including zero
df.head()      # No changes in original DataFrame
```

```python
a = df.drop(0) 
a.head()
```

```python
# Drop by label

df.drop(0, inplace = True) # if i want to modify and want to see changes in original DataFrame, then this command
df.head()
```

```python
df.drop(0, inplace = True)
df.head()
```

```python
df.drop(3, inplace = True)
df.head()
```
