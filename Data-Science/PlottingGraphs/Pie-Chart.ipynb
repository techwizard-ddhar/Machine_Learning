```python
import matplotlib.pyplot as plt
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
sizes=[3,4,6,2]
colors = ["blue", "purple", "red", "pink"]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, colors=colors, labels=labels, autopct="%.1f%%") # autopct --> import matplotlib.pyplot as plt

plt.show()
```

```python
# explode --> wedge is separated from the rest of the pie chart.

explode = [0.25,0.1,0,0]
plt.pie(sizes, colors=colors, labels=labels, autopct="%.1f%%", explode=explode)
plt.show()
```
