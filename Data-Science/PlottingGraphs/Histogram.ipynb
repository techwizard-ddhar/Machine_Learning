```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,1,3,4,2,1,1]

plt.hist(a)
plt.show()
```

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]

plt.hist(b, edgecolor = "black")
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
import numpy as np

x = np.arange(23)
plt.hist(b, edgecolor = "black")
plt.xticks(x)
plt.show()
```

```python
plt.hist(b, edgecolor = "black", bins = 21)  # bins-->number of intervals
plt.xticks(x)
plt.show()
```

# [1,2] --> 1 is included but not 2
# [2,3] --> 2 is included but not 3
   .
   .
   .
# [20,21] --> 20 is included but not 21
# [21,22] --> both 21 and 22 is included
