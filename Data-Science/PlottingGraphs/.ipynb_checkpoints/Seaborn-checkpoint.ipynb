```python
import seaborn as sns
```

```python
flight = sns.load_dataset("flights")
```

```python
flight
```

```python
data = flight.pivot_table(index="month", columns="year", values="passengers")
```

```python
data
```

```python
sns.heatmap(data)
```

# Pair-plot

```python
sns.pairplot(flight)
```

```python
sns.scatterplot(x='month', y='year', hue='passengers', data=flight)
```
