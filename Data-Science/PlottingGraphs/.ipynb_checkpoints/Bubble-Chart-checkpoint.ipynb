```python
import matplotlib.pyplot as plt
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]

plt.scatter(year,salary, s=population, c="red") # s = size

plt.plot(year, salary, linestyle='--', color="blue")
plt.title("Bubble-chart")
plt.xlabel("year")
plt.ylabel("salary")
plt.grid(True)
plt.show()
```

```python
import numpy as np
np.arange(len(year))
```

```python
c = np.arange(len(year)) # gets color based on size

plt.scatter(year,salary, s=population, c=c, alpha=0.5, edgecolor="red", marker="^") 

plt.text(year[1], salary[1], "120") # text overlap
plt.show()
```

```python
c = np.arange(len(year)) # gets color based on size

plt.scatter(year,salary, s=population, c=c, alpha=0.5, edgecolor="red", marker="^") 

plt.text(year[1] + 0.2, salary[1] + 0.1, population[1]) # overcome overlap

plt.show()
```

```python
c = np.arange(len(year)) # gets color based on size

plt.scatter(year,salary, s=population, c=c, alpha=0.5, edgecolor="red", marker="^") 

for i in range(len(population)):
    plt.text(year[i] + 0.2, salary[i] + 0.5, population[i]) # overcome overlap

plt.show()
```
